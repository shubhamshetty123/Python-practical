```python
class person:

  def __init__(self,name,age):

    self.name = name
    self.age = age

  def introduce(self):
    print("My name is:- ",self.name," and my age is:- ",self.age)

person1 = person("Alice",25)
person1.introduce()
```

```python
def multiply(a,b):
  return a*b

result = multiply(3,4)
print(result)
```

```python
class place:
  def __init__(self,city,state):

    self.city = city
    self.state = state

  def intro(self):
    print("I live in ",self.city," which is in ",self.state)

place1 = place("Mumbai","Maharashtra")
place1.intro()
```
